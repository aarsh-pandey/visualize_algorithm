# Sorting Algorithm

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import time
import sys
sys.setrecursionlimit(100000000)
np.random.seed(101)
```

```python
input_size = (np.arange(1,1001,5))

# input_size
```

### Selection Sort

```python
def SelectionSort(array):
    
    for i in range(len(array)):
        
        for j in range(i+1,len(array)):    
            if(array[j]<array[i]):
                # swaping values
                array[i],array[j]=array[j],array[i]
                
```

```python
def SelectionSortForRandomArray(size_of_array):
    array = np.random.randint(-10000,10000,size=size_of_array)
    t0=time.time()
    SelectionSort(array)
    t1=time.time() - t0
    return t1
```

```python
plt.plot(input_size,np.array([SelectionSortForRandomArray(i) for i in input_size]),color='#ff0000',label='Selection Sort')
plt.title('Selection Sort')
plt.xlabel('Input Size')
plt.ylabel('Time Required')
plt.legend()
```

### Bubble Sort

```python
def BubbleSort(array):
    
    for _ in range(len(array)):
        for i in range(len(array)-1):
            if(array[i]>array[i+1]):
                #Swaping values
                array[i],array[i+1]=array[i+1],array[i]
```

```python
def BubbleSortForRandomArray(size_of_array):
    array = np.random.randint(-10000,10000,size=size_of_array)
    t0=time.time()
    BubbleSort(array)
    t1=time.time() - t0
    return t1
```

```python
plt.plot(input_size,np.array([BubbleSortForRandomArray(i) for i in input_size]),color='#00ff00',label='Bubble Sort')
plt.title('Bubble Sort')
plt.xlabel('Input Size')
plt.ylabel('Time Required')
plt.legend()
```

### Quick Sort

```python
def Quick_Sort_Partition(array,low_index,high_index):
    pivot_index = np.random.randint(low_index,high_index+1)
    pivot_value = array[pivot_index]
    
    #Swaping values
    array[high_index],array[pivot_index] = array[pivot_index],array[high_index]
    
    i=low_index
    
    for j in range(low_index,high_index):
        
        if array[j]<pivot_value:
            #Swaping values
            array[i],array[j]=array[j],array[i]
            i+=1
    array[i],array[high_index]=array[high_index],array[i]
    pivot_index = i
    return pivot_index
    
    
def QuickSort(array=None,low_index=None,high_index=None):
    if(low_index==None):
        low_index=0
        
    if(high_index==None):
        high_index = len(array)-1
        
    if(low_index<high_index):
        
        pivot_index = Quick_Sort_Partition(array,low_index,high_index)
        
        QuickSort(array=array,low_index=low_index,high_index=pivot_index-1)
        QuickSort(array=array,low_index=pivot_index+1,high_index=high_index)
    
```

```python
def QuickSortForRandomIntArray(size_of_array):
    array = np.random.randint(-10000,10000,size=size_of_array)
    t0=time.time()
    QuickSort(array)
    t1=time.time()-t0
    return t1
```

```python
plt.plot(input_size,np.array([QuickSortForRandomIntArray(i) for i in input_size]),color='#0000ff',label='Quick Sort')
plt.title('Quick Sort')
plt.xlabel('Input Size')
plt.ylabel('Time Required')
plt.legend()
```

### Merge Sort

```python
def MergeArray(array,low_index,mid_index,high_index):
    
    left_sub_array = np.copy(array[low_index:mid_index+1])
    right_sub_array = np.copy(array[mid_index+1:high_index+1])
    
    i=j=0
    k=low_index
    
    while i<len(left_sub_array) and j<len(right_sub_array):
        
        if(left_sub_array[i]<right_sub_array[j]):
            array[k]=left_sub_array[i]
            i+=1
        else:
            array[k]=right_sub_array[j]
            j+=1
            
        k+=1
        
    while i<len(left_sub_array):
        array[k]=left_sub_array[i]
        i+=1
        k+=1
    while j<len(right_sub_array):
        array[k]=right_sub_array[j]
        j+=1
        k+=1
        
def MergeSort(array,low_index=None,high_index=None):
    if(low_index==None):
        low_index=0
    if(high_index==None):
        high_index=len(array)-1
        
    if low_index<high_index:
        mid_index = (low_index+high_index)//2
        
        MergeSort(array,low_index,mid_index)
        MergeSort(array,mid_index+1,high_index)
        
        MergeArray(array,low_index,mid_index,high_index)
```

```python
def MergeSortForRandomIntArray(size_of_array):
    array = np.random.randint(-10000,10000,size=size_of_array)
    t0=time.time()
    MergeSort(array)
    t1=time.time()-t0
    return t1
```

```python
plt.plot(input_size,np.array([MergeSortForRandomIntArray(i) for i in input_size]),color='#0000ff',label='Merge Sort')
plt.title('Merge Sort')
plt.xlabel('Input Size')
plt.ylabel('Time Required')
plt.legend()
```
